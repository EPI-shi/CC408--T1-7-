```python
# Importamos todas las librerías para los códigos siguientes
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PowerTransformer
from scipy.stats import pearsonr
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Resolver acá

# Subimos el archivo manualmente desde la computadora a Colab
uploaded = files.upload()

# Cargamos el archivo CSV en un DataFrame
file_name = "Base Airbnb NY.csv"
airbnb_data = pd.read_csv(file_name)

# Mostrar las primeras filas para confirmar la carga
print(airbnb_data.head())


# 1. Tengan en cuenta si hay valores duplicados y eliminénlos
# Eliminamos duplicados
airbnb_data_cleaned = airbnb_data.drop_duplicates()
```

```python
# 4. Si hay observaciones con outliers o valores que no tienen sentido, tomen una decision. Expliquen las decisiones tomadas.
# Empezamos por graficar todas las variables que nos quedan (las columnas), menos 'id', para ver si hay que aplicar alguna transformación

# Tomamos las columnas que queremos graficar
columns_to_plot = ['neighbourhood_group', 'latitude', 'longitude', 'room_type', 'price',
                   'minimum_nights', 'number_of_reviews', 'reviews_per_month',
                   'calculated_host_listings_count', 'availability_365']

# Configuramos el tamaño de los gráficos
plt.figure(figsize=(16, 20))

# Iteramos por cada columna y creamos su gráfico de distribución
for i, column in enumerate(columns_to_plot, 1):
  plt.subplot(5, 2, i)  # Dividimos la figura en una grilla de 5 filas y 2 columnas
  if airbnb_data_cleaned[column].dtype == 'object' or column in ['neighbourhood_group', 'room_type']:
    # Si la columna es categórica (texto) usamos un countplot
    sns.countplot(data=airbnb_data_cleaned, x=column, color='purple')
  else:
    # Si la columna es numérica, usamos un histplot
    sns.histplot(data=airbnb_data_cleaned, x=column, color='purple', bins=30, kde=True)

  plt.title(f'Distribución de {column}')
  plt.tight_layout()  # Acomodamos el layout para que no se solapen los gráficos

# Mostramos los gráficos
plt.show()
```

```python
# Continuamos por transformar algunas variables

# TRANSFORMACIONES LOGARITMICAS

# PRICE --> TRANSFORMACION LOGARITMICA

# Transformamos la columna 'price' a logaritmo natural
airbnb_data_cleaned['log_price'] = np.log(airbnb_data_cleaned['price'])

# Graficamos la distribución de 'price'
plt.figure(figsize=(8, 6))
sns.histplot(data=airbnb_data_cleaned, x='log_price', color='purple', bins=30, kde=True)
plt.title('Distribución de Price (Transformación Logarítmica)')
plt.xlabel('Ln')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Logramos que la transformación log haga que price tenga una distribución más simétrica

# NUMBER OF REVIEWS --> TRANSFORMACION LOGARITMICA

# Transformamos la columna 'number of reviews' a logaritmo natural
airbnb_data_cleaned['log_number_of_reviews'] = np.log(airbnb_data_cleaned['number_of_reviews'])

# Graficamos la distribución de 'number of reviews'
plt.figure(figsize=(8, 6))
sns.histplot(data=airbnb_data_cleaned, x='log_number_of_reviews', color='purple', bins=30, kde=True)
plt.title('Distribución de Number of Reviews (Transformación Logarítmica)')
plt.xlabel('Ln')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Logramos que la transformación log haga que number of reviews tenga una distribución más simétrica


# REVIEWS PER MONTH --> TRANSFORMACION LOGARITMICA

# Transformamos la columna 'reviews_per_month' a logaritmo natural
airbnb_data_cleaned['log_reviews_per_month'] = np.log(airbnb_data_cleaned['reviews_per_month'])

# Graficamos la distribución de 'reviews_per_month'
plt.figure(figsize=(8, 6))
sns.histplot(data=airbnb_data_cleaned, x='log_reviews_per_month', color='purple', bins=30, kde=True)
plt.title('Distribución de Reviews per Month (Transformación Logarítmica)')
plt.xlabel('Ln')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Logramos que la transformación log haga que reviews per month tenga una distribución más simétrica


# Si bien number of reviews y reviews per month mejoraron con la transformación logarítmica, al intentar filtrarles outliers no nos era conveniente
# utilizar la transformación porque se iban muchos datos
```

```python
# Detectamos los outliers usando el método del rango intercuartílico (IQR) con factor +/- 3
def detectar_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Función para filtrar outliers
def filtrar_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Variables a analizar
variables = ['latitude', 'longitude', 'log_price', 'reviews_per_month']

# Identificamos y mostramos los outliers para cada variable
outliers = {}
for var in variables:
    outliers[var] = detectar_outliers_iqr(airbnb_data_cleaned, var)
    print(f"Outliers en {var}: {len(outliers[var])} observaciones")

# Graficamos boxplots antes de eliminar outliers
for var in variables:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=airbnb_data_cleaned[var], palette='rocket', showfliers=True)  # Eje Y
    plt.title(f'Boxplot de {var} antes de eliminar outliers')
    plt.ylabel(var)
    plt.xlabel('Valores')
    plt.show()

# Filtramos el dataset original para remover los outliers identificados
for var in variables:
    airbnb_data_cleaned = filtrar_outliers(airbnb_data_cleaned, var)

# Mostrar el número de observaciones restantes después de la limpieza
print(f"Observaciones restantes después del filtrado: {len(airbnb_data_cleaned)}")
```

```python
# 5.Transformen las variables ‘neighbourhood_group’ y ‘room_type’ a variables numericas.

airbnb_data_cleaned['neighbourhood_group_num'] = airbnb_data_cleaned['neighbourhood_group'].astype('category').cat.codes
airbnb_data_cleaned['room_type_num'] = airbnb_data_cleaned['room_type'].astype('category').cat.codes

# Para ver a qué número equivale cada barrio
print(airbnb_data_cleaned[['neighbourhood_group', 'neighbourhood_group_num']].drop_duplicates())

# Para ver a qué número equivale cada tipo de depto
print(airbnb_data_cleaned[['room_type', 'room_type_num']].drop_duplicates())
```

```python
# 6. Con la ayuda de los comandos groupby y join o merge, creen una columna que tenga la cantidad de oferentes por
# “Neighbourhood group”. Llamen a esa nueva columna offer group

# Creamos la columna 'offer_group' que cuenta la cantidad de oferentes por 'neighbourhood_group'
offer_group = airbnb_data_cleaned.groupby('neighbourhood_group')['id'].transform('count')
airbnb_data_cleaned['offer_group'] = offer_group

# Vemos en 'offer group' la cantidad de oferentes que hay por barrio.
print(offer_group)

# Como el df es muy largo, creamos un resumen para ver cuántos oferentes hay por barrio una sola vez:
offer_group_summary = airbnb_data_cleaned.groupby('neighbourhood_group')['id'].count()
print(offer_group_summary)
```

```python
# Seleccionamos las columnas de interés para el PCA
columns_of_interest = [
    'neighbourhood_group_num', 'latitude', 'longitude',
    'room_type_num', 'price', 'minimum_nights',
    'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365'
]

# Filtramos el DataFrame original para trabajar solo con estas columnas
airbnb_subset = airbnb_data_cleaned[columns_of_interest]

  # Estandarización de las variables
scaler = StandardScaler(with_std=True, with_mean=True)
airbnb_transformed = pd.DataFrame(scaler.fit_transform(airbnb_subset), columns=airbnb_subset.columns)

  # Ajustamos el modelo PCA con todas las variables escaladas
pca = PCA(n_components=2)  # Especificamos que queremos dos componentes
airbnb_pca = pca.fit_transform(airbnb_transformed)

  # Mostramos la varianza explicada por los dos primeros componentes
print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)
print(f"Porcentaje de varianza explicada con los dos primeros componentes: {pca.explained_variance_ratio_[:2].sum() * 100:.2f}%")

  # Creamos un DataFrame con los scores de los componentes principales
  # Agregamos las columnas de 'room_type_num' y 'neighbourhood_group_num' al DataFrame
scores_df = pd.DataFrame(airbnb_pca, columns=['PC1', 'PC2'])
scores_df['room_type_num'] = airbnb_subset['room_type_num']
scores_df['neighbourhood_group_num'] = airbnb_subset['neighbourhood_group_num']

  #Graficamos las observaciones en el espacio de los dos primeros componentes principales
plt.figure(figsize=(10, 6))
  #Usamos seaborn para colorear por 'room_type_num' y estilo por 'neighbourhood_group_num'
sns.scatterplot(x='PC1', y='PC2', data=scores_df, hue='room_type_num', style='neighbourhood_group_num', alpha=0.7, s=50)

plt.title('Gráfico de PCA en dos dimensiones (PC1 y PC2) con Color por Tipo de Habitación y Estilo por Barrio')
plt.xlabel('Componente Principal 1 (PC1)')
plt.ylabel('Componente Principal 2 (PC2)')
plt.grid()
plt.show()
```

#### Ejercicio 8
Partan la base en una base de prueba (test) y una de entrenamiento (train) utilizando el comando <code>train_test_split</code>. La base de entrenamiento debe comprender el 70% de los datos, y la semilla a utilizar (*random state instance*) debe ser 201. Establezca a price como su variable dependiente en la base de entrenamiento (vector y). El resto de las variables ser´an las variables independientes (matriz X). Recuerden agregar la columna de unos (1).

```python
# Resolver acá
# Selección de columnas relevantes
variables_interes = [
    'neighbourhood_group_num', 'latitude', 'longitude',
    'room_type_num', 'minimum_nights', 'number_of_reviews',
    'reviews_per_month', 'calculated_host_listings_count', 'availability_365'
]

# Definimos la variable dependiente 'price' y las variables independientes en el DataFrame
X = airbnb_data_cleaned[variables_interes]  # Matriz X con las variables independientes
y = airbnb_data_cleaned['log_price']            # Vector y con la variable dependiente

# Dividir en conjuntos de entrenamiento y prueba (70% para entrenamiento y 30% para prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=201)

# Agregar una columna de unos a la matriz X_train
X_train = X_train.copy()  # Creamos una copia para no modificar el original
X_train['Intercept'] = 1  # Añadir columna de unos
# Agregar la columna Intercept con el mismo valor que en X_train (1) para que coincidan
X_test['Intercept'] = 1

# Mostrar las dimensiones de las bases resultantes
print(f"Dimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print(f"Dimensiones de y_train: {y_train.shape}")
print(f"Dimensiones de y_test: {y_test.shape}")

# Visualización de las primeras filas de la base de entrenamiento con la nueva columna
X_train.head()
```

#### Ejercicio 9
Implementen una regresión lineal y comenten los resultados obtenidos.

```python
# Resolver acá

# Crear el modelo y ajustarlo con la base de entrenamiento
model = LinearRegression()
model.fit(X_train, y_train)

# Realizar predicciones con la base de prueba
y_pred = model.predict(X_test)

# Calcular R2 para ver la calidad del ajuste
r2_train = model.score(X_train, y_train)
r2_test = model.score(X_test, y_test)

print(f"R² para el conjunto de entrenamiento: {r2_train}")
print(f"R² para el conjunto de prueba: {r2_test}")

# Mostrar los coeficientes y el intercept del modelo
print(f"Intercepto del modelo: {model.intercept_}")
print(f"Coeficientes de las variables: {model.coef_}")

# Comparación gráfica de los valores predichos vs. reales
plt.scatter(y_test, y_pred, color='pink', alpha=0.6)
plt.xlabel("Valores Reales")
plt.ylabel("Valores Predichos")
plt.title("Comparación entre valores reales y predichos")
plt.show()
```
