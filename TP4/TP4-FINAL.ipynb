```python
# Notamos que hay columnas que se repiten, algunas terminan en "_x" y otras en "_y". Esto ocurre cuando se mergean. Vamos
# a ver si contienen los mismos valores, quizás es redundante quedarnos con ambas

# Listar columnas con sufijos '_x' y '_y'
columnas_x = [col for col in base_unificada.columns if col.endswith('_x')]
columnas_y = [col for col in base_unificada.columns if col.endswith('_y')]

# Comparar valores entre columnas '_x' y '_y' y decidir qué conservar
for col_x in columnas_x:
    col_base = col_x[:-2]  # Nombre base de la columna
    col_y = col_base + '_y'
    if col_y in base_unificada.columns:
        # Si las columnas son idénticas, conservar una sola
        if base_unificada[col_x].equals(base_unificada[col_y]):
            base_unificada[col_base] = base_unificada[col_x]
        else:
            # Si no son iguales, combinar tomando los valores no nulos
            base_unificada[col_base] = base_unificada[col_x].combine_first(base_unificada[col_y])
        # Eliminar las columnas '_x' y '_y'
        base_unificada.drop(columns=[col_x, col_y], inplace=True)
    else:
        # Renombrar la columna '_x' al nombre base
        base_unificada.rename(columns={col_x: col_base}, inplace=True)
```

```python
# VARIABLE 1: Dependencia económica (proporción de personas fuera de edad de trabajar por hogar)

# Como en la limpieza eliminamos ix_men10, calculamos la cantidad de menores de 10 años restándole ix_mayeq10 (mayores de 10)
# al total de integrantes por hogar

# Calculamos la dependencia económica
base_unificada_limpia2['dependencia_economica'] = (
    (base_unificada_limpia2['ix_tot'] - base_unificada_limpia2['ix_mayeq10']) / base_unificada_limpia2['ix_tot']
)

# Verificamos valores únicos
print(base_unificada_limpia2['dependencia_economica'].unique())
```

```python
# VARIABLE 2: Proporción de personas sin secundaria completa

# Filtramos las personas que no han completado la secundaria
# 1 = Primaria Incompleta (incluye educación especial)
# 2 = Primaria Completa
# 3 = Secundaria Incompleta
# 7 = Sin instrucción
# 9 = Ns./Nr.
sin_secundaria = base_unificada_limpia2[base_unificada_limpia2['nivel_ed'].isin([1, 2, 3, 7, 9])]

# Calculamos la cantidad de personas sin secundaria completa por hogar (codusu y nro_hogar)
sin_secundaria_count = (
    sin_secundaria.groupby(['codusu', 'nro_hogar']).size()
    .reset_index(name='sin_secundaria_count')
)

# Calculamos la cantidad total de personas por hogar (codusu y nro_hogar)
total_personas_hogar = (
    base_unificada_limpia2.groupby(['codusu', 'nro_hogar'])['ix_tot'].sum()
    .reset_index(name='total_personas_hogar')
)

# Combinamos ambos resultados en la base principal
base_unificada_limpia2 = base_unificada_limpia2.merge(
    sin_secundaria_count, on=['codusu', 'nro_hogar'], how='left'
).merge(
    total_personas_hogar, on=['codusu', 'nro_hogar'], how='left'
)

# Calculamos la proporción de personas sin secundaria completa
base_unificada_limpia2['proporcion_sin_secundaria'] = (
    base_unificada_limpia2['sin_secundaria_count'] / base_unificada_limpia2['total_personas_hogar']
)

# Visualizamos el resultado
print(base_unificada_limpia2[['codusu', 'nro_hogar', 'proporcion_sin_secundaria']].dropna())
```

```python
# VARIABLE 3: Proporción de subsidios sobre el ingreso total del hogar:

# Calculamos la proporción de subsidios sobre el ingreso total del hogar
base_unificada_limpia2['proporcion_subsidios_ingreso'] = base_unificada_limpia2['v5_m'] / base_unificada_limpia2['itf']

# Visualizamos el resultado
print(base_unificada_limpia2['proporcion_subsidios_ingreso'].unique()) # Para chequear que nos den proporciones de 0 a 1
```

```python
#entrenamos modelos de regresión logística con L1 (LASSO) y L2 (Ridge)
def train_and_evaluate(X_train, X_test, y_train, y_test, penalty, year):
    # Crear y entrenar el modelo
    model = LogisticRegression(penalty=penalty, solver='liblinear', C=1, random_state=101)
    model.fit(X_train, y_train)

    # Predecir resultados
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Evaluar métricas
    cm = confusion_matrix(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    accuracy = accuracy_score(y_test, y_pred)

    # Mostrar resultados
    print(f"Resultados para {penalty.upper()} - Año {year}:")
    print("Matriz de Confusión:")
    print(cm)
    print(f"AUC: {auc:.2f}, Accuracy: {accuracy:.2f}")

    # Graficar curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.figure()
    plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc:.2f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f"Curva ROC - {penalty.upper()} Año {year}")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# Evaluamos para 2004
train_and_evaluate(X_train_2004, X_test_2004, y_train_2004, y_test_2004, penalty='l1', year=2004)  # LASSO
train_and_evaluate(X_train_2004, X_test_2004, y_train_2004, y_test_2004, penalty='l2', year=2004)  # Ridge

# Evaluamos para 2024
train_and_evaluate(X_train_2024, X_test_2024, y_train_2024, y_test_2024, penalty='l1', year=2024)  # LASSO
train_and_evaluate(X_train_2024, X_test_2024, y_train_2024, y_test_2024, penalty='l2', year=2024)  # Ridge
```

```python
 # creamos la función para calcular proporción de variables ignoradas por LASSO
def calculate_ignored_proportion(X, y, lambdas):
    ignored_proportions = []
    for c in lambdas:
        lasso_model = LogisticRegression(penalty='l1', solver='liblinear', C=c, random_state=101)
        lasso_model.fit(X, y)
        ignored_proportions.append(np.mean(lasso_model.coef_ == 0))
    return ignored_proportions

# Calculamos proporción para LASSO
ignored_proportions = calculate_ignored_proportion(X_train_2004, y_train_2004, lambdas)

# Graficamos la proporción de variables ignoradas
plt.figure(figsize=(12, 6))
sns.lineplot(x=np.log10(lambdas), y=ignored_proportions)
plt.title('Proporción de variables ignoradas por LASSO (2004)')
plt.xlabel('Log(λ)')
plt.ylabel('Proporción de variables ignoradas')
plt.show()
```
