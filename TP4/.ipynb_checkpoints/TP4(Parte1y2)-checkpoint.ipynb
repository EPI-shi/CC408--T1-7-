```python
# Importamos las librerías necesarias
import pandas as pd
```

```python
# Cargamos bases hogar
base_hogar_2004 = pd.read_stata("Hogar_t104.dta")
base_hogar_2024 = pd.read_excel("usu_hogar_T124.xlsx")
```

```python
# Cargamos bases individual
base_individual_2004 = pd.read_stata('Individual_t104.dta')
base_individual_2024 = pd.read_excel('usu_individual_T124.xlsx')
```

```python
# Uniformizamos nombres de columnas
base_hogar_2004.columns = base_hogar_2004.columns.str.lower()
base_hogar_2024.columns = base_hogar_2024.columns.str.lower()

base_individual_2004.columns = base_individual_2004.columns.str.lower()
base_individual_2024.columns = base_individual_2024.columns.str.lower()
```

```python
# Cambiamos las etiquetas de 'aglomerado' en base_hogar_2004 y base_individual_2004
map_aglomerados = {
    'Ciudad de Buenos Aires': 32,
    'Partidos del GBA': 33
}
base_hogar_2004['aglomerado'] = base_hogar_2004['aglomerado'].replace(map_aglomerados)
base_individual_2004['aglomerado'] = base_individual_2004['aglomerado'].replace(map_aglomerados)

# Filtramos por aglomerados CABA y GBA
aglomerados_valid = [32, 33]
base_hogar_2004 = base_hogar_2004[base_hogar_2004['aglomerado'].isin(aglomerados_valid)]
base_hogar_2024 = base_hogar_2024[base_hogar_2024['aglomerado'].isin(aglomerados_valid)]

base_individual_2004 = base_individual_2004[base_individual_2004['aglomerado'].isin(aglomerados_valid)]
base_individual_2024 = base_individual_2024[base_individual_2024['aglomerado'].isin(aglomerados_valid)]
```

```python
# Realizamos los merges entre bases individuales y de hogares por año
base_respondieron_2004 = pd.merge(
    base_individual_2004, base_hogar_2004, on=['codusu', 'nro_hogar'], how='inner'
)
base_respondieron_2024 = pd.merge(
    base_individual_2024, base_hogar_2024, on=['codusu', 'nro_hogar'], how='inner'
)
```

```python
# Concatenamos las bases de 2004 y 2024
base_unificada = pd.concat([base_respondieron_2004, base_respondieron_2024])
base_unificada # Nos queda una base unificada de 14698 rows × 268 columns

# Guardamos la base unificada en un csv
base_unificada.to_csv("base_unificada.csv", index=False)
```

```python
# Notamos que hay columnas que se repiten, algunas terminan en "_x" y otras en "_y". Esto ocurre cuando se mergean. Vamos
# a ver si contienen los mismos valores, quizás es redundante quedarnos con ambas

# Listar columnas con sufijos '_x' y '_y'
columnas_x = [col for col in base_unificada.columns if col.endswith('_x')]
columnas_y = [col for col in base_unificada.columns if col.endswith('_y')]

# Comparar valores entre columnas '_x' y '_y' y decidir qué conservar
for col_x in columnas_x:
    col_base = col_x[:-2]  # Nombre base de la columna
    col_y = col_base + '_y'
    if col_y in base_unificada.columns:
        # Si las columnas son idénticas, conservar una sola
        if base_unificada[col_x].equals(base_unificada[col_y]):
            base_unificada[col_base] = base_unificada[col_x]
        else:
            # Si no son iguales, combinar tomando los valores no nulos
            base_unificada[col_base] = base_unificada[col_x].combine_first(base_unificada[col_y])
        # Eliminar las columnas '_x' y '_y'
        base_unificada.drop(columns=[col_x, col_y], inplace=True)
    else:
        # Renombrar la columna '_x' al nombre base
        base_unificada.rename(columns={col_x: col_base}, inplace=True)
```

```python
# EJERCICIO 4. Construya variables (mínimo 3) que no estén en la base pero que sean relevantes para predecir individuos
# desocupados (por ejemplo, la proporción de personas que trabajan en el hogar).

#persona que no terminó la secundaria y recibe subsidios
#
```

```python
# VARIABLE 1: Dependencia económica (proporción de personas fuera de edad de trabajar por hogar)

# Como en la limpieza eliminamos ix_men10, calculamos la cantidad de menores de 10 años restándole ix_mayeq10 (mayores de 10)
# al total de integrantes por hogar

# Calculamos la dependencia económica
base_unificada_limpia2['dependencia_economica'] = (
    (base_unificada_limpia2['ix_tot'] - base_unificada_limpia2['ix_mayeq10']) / base_unificada_limpia2['ix_tot']
)

# Verificamos valores únicos
print(base_unificada_limpia2['dependencia_economica'].unique())

# Por qué sería relevante esta variable?:
# La dependencia económica puede predecir la desocupación al reflejar la carga sobre los miembros ocupados del hogar,
# asociándose con contextos de vulnerabilidad socioeconómica y menor acceso al mercado laboral.
```

```python
# VARIABLE 2: Proporción de personas sin secundaria completa

# Filtramos las personas que no han completado la secundaria
# 1 = Primaria Incompleta (incluye educación especial)
# 2 = Primaria Completa
# 3 = Secundaria Incompleta
# 7 = Sin instrucción
# 9 = Ns./Nr.
sin_secundaria = base_unificada_limpia2[base_unificada_limpia2['nivel_ed'].isin([1, 2, 3, 7, 9])]

# Calculamos la cantidad de personas sin secundaria completa por hogar (codusu y nro_hogar)
sin_secundaria_count = (
    sin_secundaria.groupby(['codusu', 'nro_hogar']).size()
    .reset_index(name='sin_secundaria_count')
)

# Calculamos la cantidad total de personas por hogar (codusu y nro_hogar)
total_personas_hogar = (
    base_unificada_limpia2.groupby(['codusu', 'nro_hogar'])['ix_tot'].sum()
    .reset_index(name='total_personas_hogar')
)

# Combinamos ambos resultados en la base principal
base_unificada_limpia2 = base_unificada_limpia2.merge(
    sin_secundaria_count, on=['codusu', 'nro_hogar'], how='left'
).merge(
    total_personas_hogar, on=['codusu', 'nro_hogar'], how='left'
)

# Calculamos la proporción de personas sin secundaria completa
base_unificada_limpia2['proporcion_sin_secundaria'] = (
    base_unificada_limpia2['sin_secundaria_count'] / base_unificada_limpia2['total_personas_hogar']
)

# Visualizamos el resultado
print(base_unificada_limpia2[['codusu', 'nro_hogar', 'proporcion_sin_secundaria']].dropna())
```

```python
# VARIABLE 3: Proporción de subsidios sobre el ingreso total del hogar:

# Calculamos la proporción de subsidios sobre el ingreso total del hogar
base_unificada_limpia2['proporcion_subsidios_ingreso'] = base_unificada_limpia2['v5_m'] / base_unificada_limpia2['itf']

# Visualizamos el resultado
print(base_unificada_limpia2['proporcion_subsidios_ingreso'].unique()) # Para chequear que nos den proporciones de 0 a 1

# La proporción de subsidios sobre el ingreso total del hogar puede ayudar a predecir la desocupación porque refleja la
# dependencia de los hogares de los subsidios sociales, lo que puede estar asociado con una menor integración al mercado
# laboral. Un hogar con alta proporción de subsidios indica posible vulnerabilidad económica y falta de empleo formal,
# lo que puede correlacionarse con mayores tasas de desocupación.
```

```python
# EJERCICIO 2. la selección del valor 𝜆 (o alpha en Python) mediante validación cruzada se realiza dividiendo
#el conjunto de entrenamiento en varios subconjuntos (folds). En cada iteración, uno de estos subconjuntos
#se reserva como conjunto de validación y el modelo se entrena en los restantes.
#Este proceso se repite para cada fold, probando diferentes valores de 𝜆
#El valor óptimo es aquel que minimiza el error promedio de validación en los folds.
#La validación cruzada permite evaluar el modelo de manera consistente en datos no utilizados durante
#el entrenamiento, maximizando la capacidad predictiva del modelo sin recurrir al conjunto de prueba.

#Por qué no usar el conjunto de prueba:
#El conjunto de prueba se reserva exclusivamente para medir la generalización del modelo en datos
#completamente nuevos, como se enfatiza en las clases. Usarlo para seleccionar 𝜆 invalidaría su propósito
#y podría llevar al sobreajuste, ya que el modelo estaría optimizado específicamente para esos datos.
#La validación cruzada, en cambio, utiliza únicamente el conjunto de entrenamiento para este ajuste,
#asegurando que el conjunto de prueba permanezca imparcial hasta la evaluación final.
```

```python
#Manejar valores faltantes Nan o infinitos
import numpy as np
from sklearn.impute import SimpleImputer

# Reemplazar valores infinitos por NaN
for dataset in [X_train_2004, X_test_2004, X_train_2024, X_test_2024]:
    dataset.replace([np.inf, -np.inf], np.nan, inplace=True)

# Imputar valores faltantes con la mediana
imputer = SimpleImputer(strategy='median')
X_train_2004 = pd.DataFrame(imputer.fit_transform(X_train_2004), columns=X_train_2004.columns)
X_test_2004 = pd.DataFrame(imputer.transform(X_test_2004), columns=X_test_2004.columns)
X_train_2024 = pd.DataFrame(imputer.fit_transform(X_train_2024), columns=X_train_2024.columns)
X_test_2024 = pd.DataFrame(imputer.transform(X_test_2024), columns=X_test_2024.columns)
```

```python
#Normalizar
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_2004 = pd.DataFrame(scaler.fit_transform(X_train_2004), columns=X_train_2004.columns)
X_test_2004 = pd.DataFrame(scaler.transform(X_test_2004), columns=X_test_2004.columns)
X_train_2024 = pd.DataFrame(scaler.fit_transform(X_train_2024), columns=X_train_2024.columns)
X_test_2024 = pd.DataFrame(scaler.transform(X_test_2024), columns=X_test_2024.columns)
```

```python
#Ahora sí, entrenar modelos de regresión logística con L1 (LASSO) y L2 (Ridge)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, accuracy_score
import matplotlib.pyplot as plt

def train_and_evaluate(X_train, X_test, y_train, y_test, penalty, year):
    # Crear y entrenar el modelo
    model = LogisticRegression(penalty=penalty, solver='liblinear', C=1, random_state=101)
    model.fit(X_train, y_train)

    # Predecir resultados
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Evaluar métricas
    cm = confusion_matrix(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    accuracy = accuracy_score(y_test, y_pred)

    # Mostrar resultados
    print(f"Resultados para {penalty.upper()} - Año {year}:")
    print("Matriz de Confusión:")
    print(cm)
    print(f"AUC: {auc:.2f}, Accuracy: {accuracy:.2f}")

    # Graficar curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.figure()
    plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc:.2f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f"Curva ROC - {penalty.upper()} Año {year}")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# Evaluar para 2004
train_and_evaluate(X_train_2004, X_test_2004, y_train_2004, y_test_2004, penalty='l1', year=2004)  # LASSO
train_and_evaluate(X_train_2004, X_test_2004, y_train_2004, y_test_2004, penalty='l2', year=2004)  # Ridge

# Evaluar para 2024
train_and_evaluate(X_train_2024, X_test_2024, y_train_2024, y_test_2024, penalty='l1', year=2024)  # LASSO
train_and_evaluate(X_train_2024, X_test_2024, y_train_2024, y_test_2024, penalty='l2', year=2024)  # Ridge
```

```python
# Cambiar a 5-fold CV
def perform_cv(X, y, penalty):
    model_cv = LogisticRegressionCV(
        Cs=lambdas,
        penalty=penalty,
        solver='liblinear',
        cv=5,  # Reducir a 5 folds
        scoring='neg_mean_squared_error',
        random_state=101
    )
    model_cv.fit(X, y)

    # Calcular MSE medio y desviación estándar
    mse_mean = -model_cv.scores_[1].mean(axis=0)  # Cambiar signo para MSE positivo
    mse_std = model_cv.scores_[1].std(axis=0)
    return mse_mean, mse_std, model_cv.C_
```

```python
 # Función para calcular proporción de variables ignoradas por LASSO
def calculate_ignored_proportion(X, y, lambdas):
    ignored_proportions = []
    for c in lambdas:
        lasso_model = LogisticRegression(penalty='l1', solver='liblinear', C=c, random_state=101)
        lasso_model.fit(X, y)
        ignored_proportions.append(np.mean(lasso_model.coef_ == 0))
    return ignored_proportions

# Calcular proporción para LASSO
ignored_proportions = calculate_ignored_proportion(X_train_2004, y_train_2004, lambdas)

# Graficar la proporción de variables ignoradas
plt.figure(figsize=(12, 6))
sns.lineplot(x=np.log10(lambdas), y=ignored_proportions)
plt.title('Proporción de variables ignoradas por LASSO (2004)')
plt.xlabel('Log(λ)')
plt.ylabel('Proporción de variables ignoradas')
plt.show()
```
