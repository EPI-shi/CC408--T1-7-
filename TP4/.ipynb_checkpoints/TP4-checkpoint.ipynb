```python
# EJERCICIO 2. 2. Descarguen la base de microdatos de la EPH correspondiente al primer trimestre de 2004 y 2024 en formato 
# .dta y .xls, respectivamente. La base de hogares se llama Hogar_t104.dta y usu_hogar_T124.xls, respectivamente. 
# Eliminen todas las observaciones que no corresponden a los aglomerados de Ciudad Autónoma de Buenos Aires o Gran Buenos 
# Aires y unan ambos trimestres en una sola base. Esto es, a la base de la encuesta individual de cada año (que usaron en el 
# TP3) unan la base de la encuesta de hogar. Asegúrese de estar usando las variables CODUSU y NRO_Hogar para el merge.
```

```python
# Importamos las librerías necesarias
import pandas as pd
```

```python
# Cargamos bases hogar
base_hogar_2004 = pd.read_stata("Hogar_t104.dta")
base_hogar_2024 = pd.read_excel("usu_hogar_T124.xlsx")
```

```python
# Cargamos bases individual
base_individual_2004 = pd.read_stata('Individual_t104.dta')
base_individual_2024 = pd.read_excel('usu_individual_T124.xlsx')
```

```python
# Uniformizamos nombres de columnas
base_hogar_2004.columns = base_hogar_2004.columns.str.lower()
base_hogar_2024.columns = base_hogar_2024.columns.str.lower()

base_individual_2004.columns = base_individual_2004.columns.str.lower()
base_individual_2024.columns = base_individual_2024.columns.str.lower()
```

```python
# Cambiamos las etiquetas de 'aglomerado' en base_hogar_2004 y base_individual_2004
map_aglomerados = {
    'Ciudad de Buenos Aires': 32,
    'Partidos del GBA': 33
}
base_hogar_2004['aglomerado'] = base_hogar_2004['aglomerado'].replace(map_aglomerados)
base_individual_2004['aglomerado'] = base_individual_2004['aglomerado'].replace(map_aglomerados)

# Filtramos por aglomerados CABA y GBA
aglomerados_valid = [32, 33]
base_hogar_2004 = base_hogar_2004[base_hogar_2004['aglomerado'].isin(aglomerados_valid)]
base_hogar_2024 = base_hogar_2024[base_hogar_2024['aglomerado'].isin(aglomerados_valid)]

base_individual_2004 = base_individual_2004[base_individual_2004['aglomerado'].isin(aglomerados_valid)]
base_individual_2024 = base_individual_2024[base_individual_2024['aglomerado'].isin(aglomerados_valid)]
```

```python
# Realizamos los merges entre bases individuales y de hogares por año
base_respondieron_2004 = pd.merge(
    base_individual_2004, base_hogar_2004, on=['codusu', 'nro_hogar'], how='inner'
)
base_respondieron_2024 = pd.merge(
    base_individual_2024, base_hogar_2024, on=['codusu', 'nro_hogar'], how='inner'
)
```

```python
# Concatenamos las bases de 2004 y 2024
base_unificada = pd.concat([base_respondieron_2004, base_respondieron_2024])
base_unificada # Nos queda una base unificada de 14698 rows × 268 columns

# Guardamos la base unificada en un csv
base_unificada.to_csv("base_unificada.csv", index=False)
```

```python
# Notamos que hay columnas que se repiten, algunas terminan en "_x" y otras en "_y". Esto ocurre cuando se mergean. Vamos
# a ver si contienen los mismos valores, quizás es redundante quedarnos con ambas

# Listar columnas con sufijos '_x' y '_y'
columnas_x = [col for col in base_unificada.columns if col.endswith('_x')]
columnas_y = [col for col in base_unificada.columns if col.endswith('_y')]

# Comparar valores entre columnas '_x' y '_y' y decidir qué conservar
for col_x in columnas_x:
    col_base = col_x[:-2]  # Nombre base de la columna
    col_y = col_base + '_y'
    if col_y in base_unificada.columns:
        # Si las columnas son idénticas, conservar una sola
        if base_unificada[col_x].equals(base_unificada[col_y]):
            base_unificada[col_base] = base_unificada[col_x]
        else:
            # Si no son iguales, combinar tomando los valores no nulos
            base_unificada[col_base] = base_unificada[col_x].combine_first(base_unificada[col_y])
        # Eliminar las columnas '_x' y '_y'
        base_unificada.drop(columns=[col_x, col_y], inplace=True)
    else:
        # Renombrar la columna '_x' al nombre base
        base_unificada.rename(columns={col_x: col_base}, inplace=True)
```

```python
# VARIABLE 1: PROPORCION DE OCUPADOS POR HOGAR
# Recordamos que en 'estado': ocupado = 1.
# ix_tot indica la cantidad de miembros por hogar

# 1. Crear una nueva columna con el total de ocupados por hogar
ocupados_por_hogar = base_unificada_limpia2[base_unificada_limpia2['estado'] == 1].groupby('nro_hogar').size()

# 2. Crear un DataFrame para el total de ocupados por hogar
ocupados_por_hogar = ocupados_por_hogar.reset_index(name='total_ocupados')

# 3. Añadir el tamaño del hogar (`ix_tot`) al DataFrame `ocupados_por_hogar`
# Usamos el primer valor de `ix_tot` de cada hogar
tamanio_hogar = base_unificada_limpia2.groupby('nro_hogar')['ix_tot'].first().reset_index()
ocupados_por_hogar = ocupados_por_hogar.merge(tamanio_hogar, on='nro_hogar', how='left')

# 4. Calcular la proporción de ocupados por hogar
ocupados_por_hogar['proporcion_ocupados_hogar'] = ocupados_por_hogar['total_ocupados'] / ocupados_por_hogar['ix_tot']

# 5. Eliminar la columna 'proporcion_ocupados_hogar' si ya existe en el DataFrame original
base_unificada_limpia2 = base_unificada_limpia2.drop(columns=['proporcion_ocupados_hogar'], errors='ignore')

# 6. Asociar esta proporción de vuelta al DataFrame original
base_unificada_limpia2 = base_unificada_limpia2.merge(
    ocupados_por_hogar[['nro_hogar', 'proporcion_ocupados_hogar']],
    on='nro_hogar',
    how='left'
)

# 7. Verificar los resultados únicos por hogar
proporcion_por_hogar = base_unificada_limpia2[['nro_hogar', 'proporcion_ocupados_hogar']].drop_duplicates()
print(proporcion_por_hogar)


```
