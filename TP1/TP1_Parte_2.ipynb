#### Ejercicio 1
Usando la API de Mercado Libre, obtener los ítems de una consulta de búsqueda. Pueden buscar cualquier producto de su interés.

```python
# Ejercicio 1

!pip install requests
import requests

#Elegimos buscar paletas de padel y definimos la variable buscar
buscar = "paletas de padel"

#Definimos la variable URL usando la api de Mercado Libre
url = "https://api.mercadolibre.com/sites/{}/search?q={}".format("MLA", buscar.replace(" ", "%20"))

# Hacemos el pedido y obtenemos la respuesta
response = requests.get(url)

# Imprimimos la respuesta
print(response.text)
```

#### Ejercicio 2
Guarden los precios de los ítems obtenidos en un dataframe y calculen el precio promedio, el mínimo y el máximo.

```python
# Ejercicio 2

!pip install pandas
import pandas as pd

# Hacemos la busqueda y guardamos el resultado en un diccionario
response = requests.get(url)
data = response.json()

# Extraemos los precios de los ítems
precios = [item["price"] for item in data["results"]]
print(precios)

# Creamos un DataFrame con los precios
df = pd.DataFrame(precios, columns=["Precio"])


# Calculamos el precio promedio, mínimo y máximo
precio_promedio = df["Precio"].mean()
precio_minimo = df["Precio"].min()
precio_maximo = df["Precio"].max()

# Imprimimos los resultados
print("Precio promedio:", precio_promedio, "ARS")
print("Precio minimo:", precio_minimo, "ARS")
print("Precio maximo:", precio_maximo, "ARS")

# Llamamos al DataFrame para ver como queda
df
```


#### Ejercicio 3
Armen un histograma de los precios. ¿Ven algún <i>outlier<i>?
Nota: pueden usar la librería de Matplotlib o la de Pandas.

```python
# Ejercicio 3

!pip install matplotlib
import matplotlib.pyplot as plt

# Creamos el histograma de los precios
plt.hist(df["Precio"], bins=20, color="pink", edgecolor="purple")
plt.title("Histograma de Precios de Paletas de Padel")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

# Para ver si hay outliers sabemos que: un valor se considera un outlier si es < que Q1−1.5 × IQR o > Q3 + 1.5 x IQR

# Calculamos Q1, Q3 y el IQR
Q1 = df["Precio"].quantile(0.25)
Q3 = df["Precio"].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identificar los outliers
outliers = df[(df["Precio"] < limite_inferior) | (df["Precio"] > limite_superior)]
print("Outliers encontrados:")
print(outliers)


#Observacion: Nos llama la tencion el gap en el mercado entre los precios 250000 y 420000
# Según el criterio que usamos de outliers, todos los artículos por encima de los $498.000 son considerados outliers.
# Sin embargo, para nosotras, debido a que son bastantes los "ouliers", simplemente las consideramos paletas muy caras.
```

#### Ejercicio 5
De la página de noticias del [diario La Nación](https://www.lanacion.com.ar/), utilicen herramientas de web scraping para obtener los links de las noticias de la portada. Guarden los links obtenidos en un dataframe y expórtenlo a un archivo de excel.

Nota 1: es posible que logren obtener los links a las noticias sin el dominio: "https://www.lanacion.com.ar/". De ser así, concatenen el dominio a la ruta del link obtenido, tal que se obtenga un link al que se pueda acceder. Es decir, que las cadenas de caracteres finales tendrán la forma: https://www.lanacion.com.ar/*texto_obtenido*)

Nota 2: junto con su entrega, adjunten una captura de la página de noticias al momento de correr su código. Eso servirá al momento de la corrección para verificar que los links obtenidos hacen referencia a las noticias de ese día y hora.

```python
# Ejercicio 5

!pip install beautifulsoup4
from bs4 import BeautifulSoup


# URL del diario La Nación
url = "https://www.lanacion.com.ar/"

# Obtenemos el contenido de la página
response = requests.get(url)
contenido = response.content

# Analizamos el contenido HTML con BeautifulSoup
soup = BeautifulSoup(contenido, "html.parser")

# Creamos una lista para almacenar los títulos
titulos = []

# Buscamos todos los elementos <h2> que contienen los títulos de las noticias
for h2 in soup.find_all("h2"):
  titulo = h2.get_text().strip()
  if titulo:
    titulos.append(titulo)

# Creamos un DataFrame con los títulos
df_titulos = pd.DataFrame(titulos, columns=["Titulo"])

# Guardamos el DataFrame en un archivo Excel
df_titulos.to_excel("titulos_noticias.xlsx", index=False)

# Vemos el dataframe
df_titulos
```
